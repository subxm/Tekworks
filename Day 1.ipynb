<h1>Local aur Global variables</h1>

```python
my_var = "heyShubham"
def hello():
    print(my_var)
hello()
```

```python
def hello():
    new_var = "byeShubham"
    print(new_var)
hello()
```

```python
x = 10
def func():
    global x
    x = x + 10

func()
print(x)
```

```python
basicSalary = int(input("Enter the Basic Salary :"))
if(basicSalary <= 10000):
    hra = basicSalary * 0.67
    da = basicSalary * 0.80
elif(basicSalary <= 20000):
    hra = basicSalary * 0.69
    da = basicSalary * 0.82
else:
    hra = basicSalary * 0.72
    da = basicSalary * 0.90
grossSalary = basicSalary + hra + da
print(grossSalary)
```

```python
salary = float(input("Enter your basic salary: "))
tax = 0
if salary <= 700000:
    tax = 0
else:
    if salary > 300000:
        tax += min(salary, 600000) - 300000 * 0.05
    if salary > 600000:
        tax += min(salary, 1200000) - 600000 * 0.15
    if salary > 1200000:
        tax += (salary - 1200000) * 0.30

print("Tax to be paid:", tax)
```
