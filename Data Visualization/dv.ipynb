## Import reqeuired libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Load the dataset

```python
df = pd.read_csv('dataset.csv')
```

```python
df
```

## 1. Histogram

```python
plt.figure(figsize=(10,6))  # set the figure size

sns.histplot(df['age'], kde=True, bins=30, color='red', edgecolor='black', alpha=0.7)
# kde = kernel density estimate is the continuous line that represents the distribution of the data
plt.title("Age Distribution")

plt.xlabel("Age")
plt.ylabel("Frequency")

plt.grid(True)
plt.xlim(0, 100)
plt.ylim(0, 40)

plt.show()
```

## 2. Scatter plot

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x='age', y='chol', data=df, color='red', edgecolor='black', alpha=0.7, s=100, marker='p')

plt.title("Age vs Cholesterol")

```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x='trestbps', y='chol', data=df, color='red', alpha=0.7, s=100, marker='o')

plt.title("Tresbps vs Cholesterol")

```

## 3. Line Plot

```python
plt.figure(figsize=(10,6))

sns.lineplot(x='age', y='chol', data=df, color='green', marker='o', alpha=0.7)

plt.title("Age vs Cholesterol (Line Plot)")

```

## 4. Box plot

### uses: outliers and range
requires only one column

```python
plt.figure(figsize=(10,6))
sns.boxplot(y='age', data=df, width=0.5, color='green')
plt.title("Age vs cholesterol (Box Plot)")
```

## 5. Pie Chart
1. Proportions
2. Distribution
3. Percentages

```python
plt.figure(figsize=(10,6))
plt.pie(
    df['target'].value_counts(),
    labels=['No Disease', 'Disease'],
    colors=['lightblue', 'salmon'],
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    textprops={'fontsize': 15},
    rotatelabels=True,
    wedgeprops={'edgecolor': 'black'},
)
```

## Bar Chart

```python
plt.figure(figsize=(10,6))
sns.barplot(x='sex', y='age', data=df, edgecolor='black',
            color='red',
            alpha=0.7,
            width=0.5)
plt.title("Average age by sex")

```
