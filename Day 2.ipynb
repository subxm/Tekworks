```python
amount = int(input("Enter withdraw amount: "))
if (amount % 100 == 0):
    if(amount >= 500):
        five = amount // 500
        amount -= five * 500
        print("500 Rupees Notes given are ->",five)
    if(amount >= 200):
        two = amount // 200
        amount -= two*200
        print("200 Rupess Notes given are ->", two)
    if(amount == 100):
        one = amount // 100
        print("100 Rupess Notes given are ->", one)

else:
    print("Please enter a valid amount which is multiples of 100.")    
```
